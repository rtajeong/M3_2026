```python
# noise removal
import re
def text_cleaner(text):
    rules = [
        {r'>\s+': '>'},  # remove one or more spaces(\s+) after a tag closes
                         # u: unicode (in Python3.0 all strings are unicode by default
        {r'\s+': ' '},  # replace consecutive spaces
        {r'\s*<br\s*/?>\s*': '\n'},  # newline after a <br> (line break, 줄바꿈)
        {r'</(div)\s*>\s*': '\n'},  # newline after </p> and </div> and <h1/>...
        {r'</(p|h\d)\s*>\s*': '\n\n'},  # newline after </p> and </div> and <h1/>...
        {r'<head>.*<\s*(/head|body)[^>]*>': ''},  # remove <head> to </head>
        {r'<a\s+href="([^"]+)"[^>]*>.*</a>': r'\1'},  # show links instead of texts
        {r'[ \t]*<[^<]*?/?>': ''},  # remove remaining tags
        {r'^\s+': ''}  # remove spaces at the beginning
    ]
    for rule in rules:
        for (k, v) in rule.items():
            regex = re.compile(k)
            text = regex.sub(v, text)
    text = text.rstrip()
    return text.lower()

text_cleaner('  The   United States of America is a \
              federal republic composed of 50 states.  ')
```

- Stemming and Lemmatization
  - Stemming: 접사(접두사, 접미사)를 단순히 잘라내어 어근(root) 형태로 변환 (예: running -> run, srudies -> studi) - 단순하고 부정확할 수 있음.
  - Lemmatization: 단어를 사전적 원형(lemma)으로 변환 (예: running -> run, studies -> study) - 정확하지만 언어학적 자원이 필요하고 느림.
  - 실제 NLP 프로젝트에서는 정확성이 중요한 경우 Lemmatization, 속도가 중요한 경우 Stemming을 선택.

```python
# stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
example_words = ["python","pythoner","pythoning","pythoned","pythonly", "studying", "taxies", "unhappiness"]
for w in example_words:
    print(ps.stem(w))
```

```python
# lemmatization: eliminating redundant prefix or suffix of a word and extract the base word (lemma)
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
example_words = ["cats", "unhappiness", "studying", "studied", "studies"]

print("\nLemmatizing example words:")
for w in example_words:
    print(f"'{w}' -> '{lemmatizer.lemmatize(w)}'")
```

- BertTokenizer: 주로 BERT 계열 모델(예: BERT, RoBERTa, DistilBERT 등)을 사용할 때 텍스트를 모델 입력 형식으로 변환하기 위해 쓰임.
- GPT 모델은 자체적으로 Byte-Pair Encoding (BPE) 또는 SentencePiece 기반 토크나이저를 사용.

```python
from transformers import BertTokenizer

# 'bert-base-uncased' 모델이 사용한 WordPiece 토크나이저를 로드한다.
# tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-cased')

# 일반 단어(tokenization)와 OOV 단어(huggingfaceization)가 포함된 예시 문장
text = "Modern NLP uses subword tokenization. Let's try huggingfaceization."

# 토큰화 실행
tokens = tokenizer.tokenize(text)
print(tokens)

# 토큰을 ID로 변환
token_ids = tokenizer.encode(text)
print("Token IDs:", token_ids)

# 다시 디코딩 (ID → 문자열)
decoded_text = tokenizer.decode(token_ids)
print("Decoded:", decoded_text)
```

```python
from transformers import GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

token_ids = tokenizer.encode(text)
print("Token IDs:", token_ids)

decoded_text = tokenizer.decode(token_ids)
print("Decoded:", decoded_text)
```
